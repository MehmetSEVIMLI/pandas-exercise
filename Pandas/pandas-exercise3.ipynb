6) horsepower'ı 155-180 arasındaki arabaların hepsini getirin.(155 dahil, 180 dahil değil)(Bu sorguyu yapabilmeniz için horsepower sütununu object durumdan float duruma getirmeniz gerekiyor.)

```python
## "?" işaretlerini "float" değere çeviremeyeceğimiz için öncelikle tüm "?" "NaN" yapıyoruz.
df["horsepower"]=df["horsepower"].replace("?", np.nan)
## ardından tüm "NaN" değerleri "float" değere çevirdik.
df["horsepower"]=df["horsepower"].astype(float)
## şimdide istenen aralıktakileri yazdıralım
df[(df["horsepower"]>155)& (df["horsepower"]<180)]

```
